```python
import pandas as pd
```

### Reading Data

```python
df = pd.read_csv("dumpling_data.csv")
df
```

### creating list of dumpling names

```python
dumpling_name = df.Name
df = df.set_index(dumpling_name)
df = df.drop(['Name'], axis=1)
df
```

### Creation of distance matrix

```python
matrix = []
for i, column_i in enumerate(dumpling_name):
    measure_list = []
    for j, column_j in enumerate(dumpling_name):
        measure_list.append(abs(df.iloc[i,:] - df.iloc[j,:]).sum())
    matrix.append(measure_list)
```

### Creation of dataframe from matrix

```python
distance_matrix = pd.DataFrame(matrix)
distance_matrix
```

### formatting index and column name

```python
distance_matrix.columns = dumpling_name
distance_matrix = distance_matrix.set_index(dumpling_name)
distance_matrix.columns.name = None
distance_matrix
```
